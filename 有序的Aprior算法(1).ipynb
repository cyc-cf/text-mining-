```python
def inn(ls,items):
    N = len(items)
    flag = 0
    for i in range(2 ** N):
        combo = []
        for j in range(N):
            if (i >> j) % 2 :
                combo.append(items[j])
        if ls == combo:
            flag += 1
        else:
            flag += 0
    if flag > 0:
        return True
    else:
        return False
```

```python
def scan(data,base):  
    dic = {}
    for i in base:
        for j in data:
            if inn(i,j):
                dic[str(i)] = dic.get(str(i),0) + 1
    return dic
```

```python
def acc_var(num):
    s = 'item[0]'
    for i in range(1,num):
        s += f'+item[{i}]'
    return s
```

```python
def get_support(data,base,min_support):
    dic = {}
    count = 0
    for i in range(1,num+1):        
        dic.update(scan(data,[eval(acc_var(i)) for item in itertools.permutations(base,i)]))
        count += 1
        print(f"第{count}次输出，结果为：{dic}")
        
    support_dic = {}
    count = 0
    for key,value in dic.items():
        support = value/num
        if support >  min_support:
            support_dic[key] =support
            count += 1 
        print(count)
    return support_dic
```

```python
def get_rlue(support_dic):
    count = 0
    for key,value in support_dic.items():
        n = len(eval(key))
        if  n == 2 :
            front = eval(key)[0]
            after = eval(key)[1]
            confi = support_dic[key]/support_dic[str([front])]
            print(f'{[front]}-->{[after]}:{confi}')
        elif n >2:
            front =  eval(key)[:n-1]
            after = eval(key)[-1]
            confi = support_dic[key]/support_dic[str(front)]
            print(f'{front}-->{[after]}:{confi}')
        count += 1
    print(count)
```
