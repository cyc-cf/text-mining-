```python
#判断候选项集是数据集的子集
def inn(ck,ls):
    #ck候选项集
    #ls  dataset的集合,eg: ls=[1,3,4]、ls=[2,3,5]、ls=[1,2,3,5]...
    N = len(ls)#子集的长度
    flag = 0
    for i in range(2 ** N): #所有的排列组合结果
        combo = []
        for j in range(N):
            if (i >> j) % 2 :
                combo.append(ls[j])
        if ck == combo:
            flag += 1
        else:
            flag += 0
    if flag > 0:
        return True
    else:
        return False
```

```python
#生成有序子集
def getArraySubSet(originArray):
    """
    :type originArray:list
    :rtype :listlist
    """
    result = [[]]
    size = len(originArray)
    for i in range(size):
        for j in range(len(result)):
            # 现有每个子集中添加新元素，作为新子集加入结果集中
            result.append(result[j]+[originArray[i]])
        # print(result)
    return result
# 测试
originArray = [1, 2, 3, 4]#从中找到有序子集
subset = getArraySubSet(originArray)
print(subset)
```

```python
#计算支持度，存放所有候选项集
import collections
import itertools

def scan_c1(data,base,min_support):
    #data 原始数据集
    #base 候选项集
    ls = []  #存放所有候选项集
    for i in range(1,len(base)+1):
        it = itertools.permutations(base,i)
        for j in it:
            ls.append(list(j))
            
    num = float(len(data))
    dic = {} #候选项集及其频次
    for i in base:
        for j in data:
            if inn(i,j): #i候选项集 j是数据集的子集
                dic[str(i)] = dic.get(str(i),0) + 1

    dic_support = {}#存放最小支持度集合
    for i in base:
        for key,value in dic.items():
            dic_support[key] = dic[key]/num
          
    return dic_support
```

```python
import collections
def scan(data,base,minSupport,k):
    #k 是候选k项集
    ls = []#存放所有子集
    for i in range(1,len(base)+1):
        it = itertools.permutations(base,i)
        for j in it:
            ls.append(list(j)) 
    num = float(len(data))
    dic = {}
    retList = []#频繁项集
    for i in itertools.permutations(base,k):
        for j in data:
            if inn(i[0],j):
                dic[str(i[0])] = dic.get(str(i[0]),0) + 1
    dic_support = {}
    for i in base:
        for key,value in dic.items():
            dic_support[key] = dic[key]/num
            if dic_support[key]>= minSupport:
                retList.append(key)
    return dic_support,retList
```
